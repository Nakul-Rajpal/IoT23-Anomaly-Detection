```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 42 seed results
accuracy = np.array([0.9998, 0.9995, 0.998, 0.9987, 0.9986, 0.9946, 0.9918, 0.9916, 0.9818, 0.7967])


dataset_size = np.array([1, 0.8, 0.6, 0.4, 0.2, 0.1, 0.05, 0.025, 0.01, 0.005])
plt.plot(dataset_size, accuracy, marker='o', label='Precision')
plt.plot(dataset_size, accuracy, marker='o', label='Accuracy')
plt.plot(dataset_size, accuracy, marker='o', label='Accuracy')
plt.xlim(0,0.4)
plt.xlabel('Dataset Size')
plt.ylabel('Precision')
plt.title('Precision vs Dataset Size: Seed 42')
plt.savefig("16_4_8_Precision.png", dpi=300)
```
